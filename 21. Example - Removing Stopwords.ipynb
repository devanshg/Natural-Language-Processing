### Subtract stopwords from 100 most common words in Alice of Wonderland

```python
import nltk
```

```python
alice = nltk.corpus.gutenberg.words('carroll-alice.txt')
```

```python
alice[:10]
```

```python
alice_norm = [word.lower() for word in alice if word.isalpha()]
```

```python
alice_fd = nltk.FreqDist(alice_norm)
```

```python
alice_100 = [word[0] for word in alice_fd.most_common(100)]
alice_100
```

```python
stopwords = nltk.corpus.stopwords.words('english')
stopwords
```

```python
set(alice_100) - set(stopwords)
```
