# For Alice in Wonderland find all parts of speech used in the words "over", "spoke" and "answer"

```python
import nltk
```

```python
alice = nltk.corpus.gutenberg.words('carroll-alice.txt')
```

```python
alice
```

```python
alice_norm = [word.lower() for word in alice if word.isalpha()]
```

```python
alice_norm.count('over')
```

```python
alice_tags = nltk.pos_tag(alice_norm, tagset='universal')
```

### Method 1

```python
alice_nouns = [word[1] for word in alice_tags if word[0] == 'over']
```

```python
nouns_fd = nltk.FreqDist(alice_nouns)
nouns_fd
```

```python
alice_nouns = [word[1] for word in alice_tags if word[0] == 'spoke']
nouns_fd = nltk.FreqDist(alice_nouns)
nouns_fd
```

```python
alice_nouns = [word[1] for word in alice_tags if word[0] == 'answer']
nouns_fd = nltk.FreqDist(alice_nouns)
nouns_fd
```

### Method 2 (Using Conditional Frequency)

```python
alice_cfd = nltk.ConditionalFreqDist(alice_tags)
```

```python
alice_cfd
```

```python
alice_cfd['over']
```

```python
alice_cfd['spoke']
```

```python
alice_cfd['answer']
```
